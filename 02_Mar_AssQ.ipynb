Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

scatter plot used to display the relationship between two variables as individual data points. It represents individual data points as dots on a two-dimensional plane, with one variable plotted on the x-axis and the other variable on the y-axis. Scatter plots are useful for identifying patterns, trends, or correlations between variables.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
```

```python
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize = (6,4))
plt.scatter(x, y)
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.title("title")
plt.grid()
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in matplotlib's pyplot module is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each with its own axes. This can be useful when you want to display multiple plots side by side or in a grid layout, making it easier to compare and analyze different data or visualizations.

```python
fig, axes = plt.subplots(2,2)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
axes[0, 0].plot(x, y)
axes[0, 0].set_title("line1")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
axes[0, 1].plot(x, y)
axes[0, 1].set_title("line2")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
axes[1, 0].plot(x, y)
axes[1, 0].set_title("line3")

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
axes[1, 1].plot(x, y)
axes[1, 1].set_title("line4")

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is type of plot used to compare different categories or groups of data. It displays data as rectangular bars, where the length of each bar represents the value of a particular group. Bar plots are commonly used to visualize and compare data across different categories, such as comparing sales by product, population by country, or grades by subjects. They provide a clear visual representation of the data, making it easy to identify patterns, trends, and comparisions between the categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize = (4, 3))
plt.bar(company, profit, color = "#3D90C7")
plt.grid()
```

```python
plt.figure(figsize = (4, 3))
plt.barh(company, profit, color = "#3D90C7")
plt.grid()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

A box plot is a type of plot used to display the distribution of a dataset. It is also known as a box-and-whisker plot. A box plot summarizes the minimum, first quartile, median, third quartile, and maximum values of a dataset. It consists of a rectangular box, which represents interquartile range, and whiskers that extend from the box to the minimum and maximum values. Box plots are useful for identifying the spread, skewness, and outliers in a dataset, making it easier to compare distributions or analyze data across different groups or categories.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

fig, axes = plt.subplots(2, 2)
axes[0, 0].boxplot(box1)
axes[1, 0].boxplot(box2)
plt.show()
```
